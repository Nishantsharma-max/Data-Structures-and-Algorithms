```python
class BSTNode:
     def __init__(self,data):
          self.data=data
          self.leftChild=None
          self.rightChild=None

def insertNode(rootNode,nodeValue):
     if rootNode.data==None:
          rootNode.data=nodeValue
     elif rootNode.data>=nodeValue:
          if rootNode.leftChild is None:
               rootNode.leftChild=BSTNode(nodeValue)
          else:
               insertNode(rootNode.leftChild,nodeValue)
     else:
          if rootNode.rightChild is None:
               rootNode.rightChild=BSTNode(nodeValue)
          else:
               insertNode(rootNode.rightChild,nodeValue)
def preOrderTransversal(rootNOde):
     if rootNOde is None:
          return 
     else:
          print(rootNOde.data)
          preOrderTransversal(rootNOde.leftChild)
          preOrderTransversal(rootNOde.rightChild)

def inorderTransversal(rootNode):
     if rootNode is None:
          return 
     else:
          inorderTransversal(rootNode.leftChild)
          print(rootNode.data)
          inorderTransversal(rootNode.rightChild)

def postOrdertransversal(rootNode):
     if rootNode is None:
          return 
     else:
          postOrdertransversal(rootNode.leftChild)
          postOrdertransversal(rootNode.rightChild)
          print(rootNode.data)

class Node:
     def __init__(self,value=None):
          self.value=value
          self.next=None
     def __str__(self):
          return str(self.value)

class Linkedlist:
     def __init__(self):
          self.Head=None
          self.Tail=None

class Queue:
     def __init__(self):
          self.queue=Linkedlist()
     
     def __str__(self):
          current=self.queue.Head
          res=""
          while current is not None:
               res+=f'{str(current.value)}  '
               current=current.next
          return res[:-2]
     def enqueue(self,val):
          node=Node(val)
          if self.queue.Head is None:
               self.queue.Head=node
               self.queue.Tail=node
          else:
               self.queue.Tail.next=node
               self.queue.Tail=node

     def isEmpty(self):
          if self.queue.Head==None:
               return True
          else:
               return False
          
     def dequeue(self):
          if not self.isEmpty():
               node=self.queue.Head
               if self.queue.Head==self.queue.Tail:
                    self.queue.Head=None
                    self.queue.Tail=None
               else:
                    self.queue.Head=self.queue.Head.next
               return node

     def peek(self):
          if not self.isEmpty():
               return self.queue.Head.value
          
     def delete(self):
          if not self.isEmpty():
               self.queue.Head=None
               self.queue.Tail=None
               return "queue deleted"



def levelOrdertransversal(rootNode):
     if rootNode is None:
          return
     else:
          customqueue=Queue()
          customqueue.enqueue(rootNode)
          while not(customqueue.isEmpty()):
               root=customqueue.dequeue()
               print(root.value.data)
               if root.value.leftChild is not None:
                    customqueue.enqueue(root.value.leftChild)
               if root.value.rightChild is not None:
                    customqueue.enqueue(root.value.rightChild)

def searchNode(rootNode,nodeValue):
     if rootNode.data==nodeValue:
          print("exist")
     elif nodeValue<=rootNode.data:
          if rootNode.leftChild is None:
               print("does't exist")
          else:
               searchNode(rootNode.leftChild,nodeValue)
     else:
          if rootNode.rightChild is None:
               print("doesn't exist")
          else:
               searchNode(rootNode.rightChild,nodeValue)

def minValueNode(bstNode):
     current=bstNode
     while current.leftChild is not None:
          current=current.leftChild
     return current


def deletenode(rootnode ,nodevalue):
     if rootnode is None:
          return rootnode
     current=rootnode
     while current.data!=nodevalue:
          parent=current
          if current.data>nodevalue:
               current=current.leftChild
          else:
               current=current.rightChild
     # delete leaf node
     if current.leftChild is None and current.rightChild is None:
          if parent.leftChild is not None:
               if parent.leftChild.data==nodevalue:
                    parent.leftChild=None
               else:
                    parent.rightChild=None
          else:
               parent.rightChild=None
     # delete node with only one child
     if current.leftChild is None and current.rightChild is not None:
          current.data=current.rightChild.data
          delete=current.rightChild
          current.rightChild=None
          return delete
     if current.leftChild is not None and current.rightChild is None:
          current.data=current.leftChild.data
          delete=current.leftChild
          current.leftChild=None
          return delete
     # delete node with two child
     if current.leftChild is not None and current.rightChild is not None:
          minnode=minValueNode(current.rightChild)
          current.data=minnode.data
          if minnode.rightChild is not None:
               minnode.data=minnode.rightChild.data
               minnode.rightChild=None
          

def deleteBST(rootnode):
     if not rootnode:
          return
     rootnode.data=None
     rootnode.leftChild=None
     rootnode.rightChild=None
     return "bst deleted"

                   





```
